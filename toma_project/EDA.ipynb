```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_all_histograms(df, figsize=(15, len(df.columns)*5)):
    # Create a figure with subplots - one for each column
    fig, axes = plt.subplots(nrows=len(df.columns), ncols=1, figsize=figsize)
    
    # If there's only one column, axes won't be an array, so we convert it
    if len(df.columns) == 1:
        axes = [axes]
    
    # Loop through each column and create appropriate visualization
    for idx, (column, ax) in enumerate(zip(df.columns, axes)):
        # Get column data type
        dtype = df[column].dtype
        
        # For numeric data (int or float)
        if pd.api.types.is_numeric_dtype(dtype):
            sns.histplot(data=df, x=column, ax=ax, kde=True)
            ax.set_title(f'Distribution of {column} (Numeric)')
            
        # For categorical/object data
        else:
            # Calculate value counts and create bar plot
            value_counts = df[column].value_counts()
            sns.barplot(x=value_counts.index, y=value_counts.values, ax=ax)
            ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
            ax.set_title(f'Distribution of {column} (Categorical)')
        
        # Add grid for better readability
        ax.grid(True, alpha=0.3)
    
    # Adjust layout to prevent overlap
    plt.tight_layout()
    # return fig

# Usage example:
plot_all_histograms(train)
```
